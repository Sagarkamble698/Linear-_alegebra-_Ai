
# 1. Vector Operations

Vectors are used in:
- Embeddings
- Recommendation Systems
- Deep Learning
- Computer Vision


```

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("Vector A:", A)
print("Vector B:", B)

addition = A + B
subtraction = A - B

print("\nVector Addition:", addition)
print("Vector Subtraction:", subtraction)
```


# 2. Dot Product

The dot product measures similarity between vectors.

### AI Applications
- Recommendation Systems
- Semantic Search
- NLP Embeddings


```

dot_product = np.dot(A, B)

print("Dot Product:", dot_product)
```


# 3. Cross Product

Cross product is mainly used in:
- 3D Graphics
- Robotics
- Physics Simulations


```

cross_product = np.cross(A, B)

print("Cross Product:", cross_product)
```


# 4. Matrix Multiplication

Matrix multiplication is the backbone of Neural Networks.

### Neural Network Formula

Output = Input × Weights


```

X = np.array([
    [1, 2],
    [3, 4]
])

W = np.array([
    [5, 6],
    [7, 8]
])

output = np.dot(X, W)

print("Input Matrix:\n", X)
print("\nWeight Matrix:\n", W)
print("\nOutput Matrix:\n", output)
```


# 5. Scalar Multiplication

Used for scaling values in mathematics and ML optimization.


```

matrix = np.array([
    [1, 2],
    [5, 6],
    [9, 0]
])

scaled_matrix = matrix * 3

print("Original Matrix:\n", matrix)
print("\nScaled Matrix:\n", scaled_matrix)
```


# 6. Matrix Transpose

Transpose is important in:
- Backpropagation
- Deep Learning
- Matrix Optimization


```

matrix = np.array([
    [1, 2],
    [3, 4]
])

transpose_matrix = matrix.T

print("Original Matrix:\n", matrix)
print("\nTranspose Matrix:\n", transpose_matrix)
```


# 7. Matrix Determinant

Determinants help determine whether a matrix is invertible.


```

matrix = np.array([
    [1, 2],
    [4, 6]
])

determinant = np.linalg.det(matrix)

print("Determinant:", determinant)
```


# 8. Inverse Matrix

Inverse matrices are used in:
- Linear Regression
- Optimization Problems
- Solving Linear Equations


```

inverse_matrix = np.linalg.inv(matrix)

print("Inverse Matrix:\n", inverse_matrix)
```


# 9. Eigenvalues & Eigenvectors

Eigen concepts are heavily used in:
- PCA (Principal Component Analysis)
- Face Recognition
- Data Compression


```

eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```


# 10. Identity Matrix

Identity matrix behaves like number 1 in matrix mathematics.


```

identity = np.eye(3)

print("Identity Matrix:\n", identity)
```


# 11. Neural Network Mathematics Example

Simple linear transformation used in Deep Learning.


```

X = np.array([
    [2, 6],
    [4, 8]
])

W = np.array([
    [0.5],
    [1]
])

y = np.dot(X, W)

print("Predictions:\n", y)
```


# 12. Data Visualization

Visualization helps understand vectors and transformations.


```

x = [0, 1]
y = [0, 2]

plt.figure(figsize=(5,5))
plt.quiver(0, 0, x[1], y[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.title("Vector Visualization")
plt.grid()
plt.show()
```


# 13. PCA Intuition

PCA reduces dimensions while preserving important information.

### Used In
- Computer Vision
- Data Compression
- Feature Engineering


```

from sklearn.decomposition import PCA

data = np.array([
    [2, 3],
    [3, 4],
    [4, 5],
    [5, 6]
])

pca = PCA(n_components=1)

reduced = pca.fit_transform(data)

print("Reduced Data:\n", reduced)
```
